```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pymysql.cursors
import time 
import os

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='asd273321',
                             db='new_media',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)


options = Options()


driver = webdriver.Chrome(os.getcwd()+'/chromedriver', chrome_options=options)
try:
    with connection.cursor() as cursor:
        driver.get('https://www.inside.com.tw/')
        sourceCode = BeautifulSoup(driver.page_source)
        article_box = sourceCode.select('div.post_list-grid_style')[0]
        articles = article_box.select('div.post_list_item')
        for article in articles:
            title = article.select('h3.post_title')[0].text
            date = article.select('li.post_date')[0].text
            tags = article.select('a.hero_slide_tag')
            #製作變數＿塞進ＴＡＧ
            tags_string = ''
            for tag in tags:
                tags_string += tag.text + ', '
            print(title)
            print(date)
            print(tags_string)
            sql = '''
            INSERT INTO `new_media`.`articles` (`title`, `date`, `tags`, `brand`)
            VALUES('{}', '{}', '{}', '{}')
            '''.format(title, date, tags_string, 'inside')
            print(sql)
            cursor.execute(sql)
            connection.commit()
        
    connection.close()
    driver.close()
except Exception as e:
    print(e)
    connection.close()
    driver.close()
```
